**Starbucks_Clustering**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage 
from matplotlib import pyplot as plt
```

```python
points=np.array([[8,1],[7,2],[7,3],[5,4],[3,3],[4,5],[3,6],[2,6],[1,5],[8,7],[5,7],[1,7],[4,8],[3,8],[7,9],[6,9],[3,9],[2,9]])
```

```python
labels = range(1, 19) 
plt.figure(figsize=(10, 7)) 
plt.subplots_adjust(bottom=0.1) 
plt.scatter(points[:,0],points[:,1], label='True Position')
```

```python
for label, x, y in zip(labels, points[:, 0], points[:, 1]): 
    plt.annotate(
    label,
    xy=(x, y), xytext=(-3, 3),
    textcoords='offset points', ha='right', va='bottom')
plt.show()
```

**Single_Link_Clustering**

```python
linked = linkage(y=points, method='single')
labellist=range(1,19)
dendrogram(linked, 
orientation='top',
labels=labellist,distance_sort='descending',show_leaf_counts=True)
plt.show()
```

**Complete_Link_Clustering**

```python
linked = linkage(points, 'complete')
labellist=range(1,19)
dendrogram(linked, 
orientation='top',
labels=labellist,distance_sort='descending',show_leaf_counts=True)
plt.show()
```

**Average_Link_Clustering**

```python
linked = linkage(points, 'average')
labellist=range(1,19)
dendrogram(linked, 
orientation='top',
labels=labellist,distance_sort='descending',show_leaf_counts=True)
plt.show()
```
